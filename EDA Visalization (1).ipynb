## Group by 

```python
df.groupby(by = 'Country')[['Sales($B)','Profit($B)','Assets($B)','Market_Value($B)']].agg([np.mean,np.median,np.max])
```

```python
df1 = df.groupby('Country')['Sales($B)'].sum().reset_index(name = 'Sales($B)')
```

```python
df1 = df1.sort_values("Sales($B)",ascending = False)
df1
```

```python
df2=df1.head(10)
df2
```

```python
sns.barplot(data = df2,x = 'Country',y = 'Sales($B)',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
```

```python
df3 = df.groupby("Company_Name")['Sales($B)'].sum().reset_index(name = 'Sales($B)')
df4 = df3.sort_values('Sales($B)',ascending = False)
df5 = df4.head(10)
sns.barplot(data = df5,x = "Company_Name",y = 'Sales($B)',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
```

```python
df3 = df.groupby("Company_Name")['Assets($B)'].sum().reset_index(name = 'Assets($B)')
df4 = df3.sort_values('Assets($B)',ascending = False)
df5 = df4.head(10)
sns.barplot(data = df5,x = "Company_Name",y = 'Assets($B)',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
```

```python
df3 = df.groupby("Company_Name")['Profit($B)'].sum().reset_index(name = 'Profit($B)')
df4 = df3.sort_values('Profit($B)',ascending = False)
df5=df4.head(10)

sns.barplot(data = df5,x = "Company_Name",y = 'Profit($B)',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
```

```python
df3 = df.groupby("Company_Name")['Market_Value($B)'].sum().reset_index(name = 'Market_Value($B)')
df4 = df3.sort_values('Market_Value($B)',ascending = False)
df5 = df4.head(10)
sns.barplot(data = df5,x = "Company_Name",y = 'Market_Value($B)',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
```

```python
df3 = df.groupby("Company_Name")['Assets($B)'].sum().reset_index(name = 'Assets($B)')
df4 = df3.sort_values('Assets($B)',ascending = False)
df5 = df4.head(10)
sns.barplot(data = df5,x = "Company_Name",y = 'Assets($B)',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
```

```python
df3 = df.groupby("Company_Name")['Ranks'].sum().reset_index(name = 'Ranks')
df4 = df3.sort_values('Ranks',ascending = True)
df5 = df4.head(10)
sns.barplot(data = df5,x = "Company_Name",y = 'Ranks',palette=sns.color_palette())
plt.xticks(rotation = 90)
plt.show()
df5
```
